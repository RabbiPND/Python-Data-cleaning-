Data Science Jobs on Glassdoor Dataset cleaning

```python
import pandas as pd
import numpy as np
import re
```

```python
df=pd.read_csv(r"C:\Users\rabbi\OneDrive\Documents\DATA ANALYSIS\Data Analystics\PYTHON\Data cleaning\Data Science Jobs on Glassdoor\Report\Data Science jobs on glassdoor.csv")
```

```python
df
```

```python
df=df.drop_duplicates()
df["Salary Estimate"] = df["Salary Estimate"].str.strip("(Glassdoor est.)").str.replace("K", "").str.replace("$", "")
df2 = df.rename(columns={"Salary Estimate": "Salary Estimate (000)"})
df["Job Description"] = df["Job Description"].str.replace("\n", ":")
def replace_multiple_colon(text):
    return re.sub(r'(:)+', ':', text)

df["Job Description"] = df["Job Description"].apply(replace_multiple_colon)
df["Company Name"] = df["Company Name"].apply(lambda x: re.sub(r'[\d\n\.\--]', '', x))
df[['City', 'State']] = df['Location'].str.split(',', n=1, expand=True)
df = df.reset_index(drop=True)
df = df.drop(columns="Location")
```
